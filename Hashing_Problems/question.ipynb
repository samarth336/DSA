# 290. Word Pattern

```python
class Solution:
    def wordPattern(self, pattern: str, s: str) -> bool:
        if len(pattern)!= len(s.split(" ")):
            return False
        s=s.split(" ")
        p_set={}
        s_set={}

        for i in range(len(pattern)):
            if pattern[i] not in p_set:
                p_set[pattern[i]]=s[i]
            if s[i] not in s_set:
                s_set[s[i]]=pattern[i]

            if p_set[pattern[i]]!=s[i] or s_set[s[i]]!=pattern[i]:
                    return False
           
        return True

s=Solution()
print(s.wordPattern("abba", "dog cat cat Fish"))
print(s.wordPattern("abba", "dog cat cat dog"))
print(s.wordPattern("aaaa", "dog dog dog dog"))
print(s.wordPattern("abba", "dog dog dog dog"))
```

# 349. Intersection of Two Arrays

normal approach

```python
class Solution:
    def intersection(self, nums1, nums2):
        nums1=set(nums1)
        nums2=set(nums2)
        result=[]
        for i in nums1:
            for j in nums2:
                if i ==j:
                    result.append(i)

        return result

s=Solution()
print(s.intersection([1,2,2,1],[2,2]))
print(s.intersection([4,9,5],[9,4,9,8,4]))
print(s.intersection([1,2,3,4,5],[6,7,8,9]))
```

optimal aproach

```python
class Solution:
    def intersection(self, nums1, nums2):
        nums1=set(nums1)
        nums2=set(nums2)
        return list(nums1.intersection(nums2))

s=Solution()
print(s.intersection([1,2,2,1],[2,2]))
print(s.intersection([4,9,5],[9,4,9,8,4]))
print(s.intersection([1,2,3,4,5],[6,7,8,9]))
```

# 350. Intersection of Two Arrays II

```python
class Solution:
    def intersect(self, nums1, nums2):
        nums1.sort()
        nums2.sort()
        i=0
        j=0
        result=[]
        while i<len(nums1) and j<len(nums2):
            if nums1[i]==nums2[j]:
                result.append(nums1[i])
                i+=1
                j+=1
            elif nums1[i]<nums2[j]:
                i+=1
            else:
                j+=1
        return result

s=Solution()
print(s.intersect([1,2,2,1],[2,2]))
print(s.intersect([4,9,5],[9,4,9,8,4]))
print(s.intersect([1,2,3,4,5],[6,7,8,9]))
```
